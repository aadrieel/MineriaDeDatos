```python

'''
Equipo: 3
Grupo: 002, Miércoles
Integrantes: 
1941502 Vidal Mosqueda Cristina
1811901 García Salazar Orlando
1644680 Macías Cortés Isaí Adriel
1849522 Mora Martínez Gerardo

Base de datos principal
Título: "US Consumer Finance Complaints"
Url: https://www.kaggle.com/cfpb/us-consumer-finance-complaints

Descripción: 
Los servicios bancarios suelen presentar fallas, por lo que a consecuencia de 
esto, los clientes presentan quejas hacia los bancos. Nuestra base de datos 
se basa en las quejas que presentan los consumidores a los bancos 
estadounidenses.

Descripción de los datos: 
Son reclamos hechos por los clientes, las cuales, son registradas por la CFPB 
(Consumer Financial Protection Bureau) que se introducen a una base de datos 
que generan una matriz de todas las quejas financieras que se hicieron en 
Estados Unidos de América.

Descripción de las columnas:
De nuestra limpieza de la base de datos, las columnas que la conforma son: 
-date_received: Fecha en la que la queja fue enviada por el cliente.
-product: El producto financiero referente a la queja que solicitó el cliente.
-issue: El título general establecido para identificar la queja del cliente.
-company: El banco en el que el cliente solicitó el servicio.
-state: Estado en el que esta localizado el banco.
-submitted_via: Medio por el cual la queja fue enviada.
-date_sent_to_company: Fecha en la que se recibió al banco el reclamo del cliente.
-company_response_to_consumer: Respuesta del banco hacia el consumidor.
-timely_response: Especifica si la queja fue contestada en el tiempo.
establecido del banco.
-consumer_disputed: Especifica si el cliente quedó satisfecho con la solución.

Justificación del uso de datos:
Nos llamó la atención los tipos de quejas y los productos financieros a los que se estaban refiriendo 
y la forma en la que las quejas fueron enviadas.
Creemos que con estos datos podemos identificar el mejor banco con menores 
quejas de este producto para ayudar al cliente a tomar una mejor decisión.

Objetivo final:
Como objetivo principal es darle al cliente las diferentes opciones para que 
así seleccione el banco adecuado dependiendo a sus necesidades, y como objetivo 
secundario el analizar que tipo de quejas tiene cada banco por interés 
informativo.

Preguntas de interés:
1) ¿Qué banco de USA presenta mayores quejas?
2) ¿Qué banco de USA presenta menores quejas?

Programación:
Título: "US Consumer Finance Complaints"
Url: https://www.kaggle.com/cfpb/us-consumer-finance-complaints
'''

import pandas as pd

#Lectura de archivo
df = pd.read_csv("consumer_complaints.csv",encoding='utf8')
#df.head()#Imprimimos los primeros 5 registros para checar si se necesita eliminar columnas

#Eliminar columnas inecesarias
aux = df.drop(['sub_product','sub_issue','consumer_complaint_narrative','company_public_response','zipcode','tags','consumer_consent_provided','complaint_id'],axis = 1)

#Checamos columnas nulas
#print(aux.isnull().any())

#Quitamos las filas con nulos, que en este caso son pocos
baselimp = aux.dropna()
#print(baselimp.isnull().any()) #Verificamos que no haya nulos

#Cambiamos los nombres de las columnas
baselimp = baselimp.rename(columns = {'date_received':'Fecha_Enviada_por_el_Cliente'})
baselimp = baselimp.rename(columns = {'product':'Producto_Financiero'})
baselimp = baselimp.rename(columns = {'issue':'Queja'})
baselimp = baselimp.rename(columns = {'company':'Banco'})
baselimp = baselimp.rename(columns = {'state':'Diminutivo_de_Estado_de_USA'})
baselimp = baselimp.rename(columns = {'submitted_via':'Medio_de_Envío'})
baselimp = baselimp.rename(columns = {'date_sent_to_company':'Fecha_Recibida_al_Banco'})
baselimp = baselimp.rename(columns = {'date_response_to_consumer':'Fecha_de_respuesta_al_cliente'})
baselimp = baselimp.rename(columns = {'timely_response':'¿Se_evaluó_a_tiempo_la_respuesta_de_la_queja?'})
baselimp = baselimp.rename(columns = {'consumer_disputed?':'¿Cliente_sastifecho_a_la_solución?'})


#Agrupamos los registros por los tipos de registros
clase_1 = baselimp[baselimp['Producto_Financiero']=='Mortgage']
clase_2 = baselimp[baselimp['Producto_Financiero']=='Debt collection']
clase_3 = baselimp[baselimp['Producto_Financiero']=='Credit reporting']
clase_4 = baselimp[baselimp['Producto_Financiero']=='Credit card']
clase_5 = baselimp[baselimp['Producto_Financiero']=='Bank account or service']
clase_1.head()#Checamos que las agrupaciones sean las correctas

#Exportamos las agrupaciones en csv
clase_1.to_csv('Mortgage.csv', index=False)
clase_2.to_csv('Debt collection.csv', index=False)
clase_3.to_csv('Credit reporting.csv', index=False)
clase_4.to_csv('Credit Card.csv', index=False)
clase_5.to_csv('Bank account or service.csv', index=False)
```
