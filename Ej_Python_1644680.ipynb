```python
import math as m

#Ejercicio 1
"Realiza una variable con tu matricula y realiza una secuencia de imprimir con tu nombre y tu matricula concatenados."

mat = 1644680
name = "Isaí Adriel Macías Cortés"

print (name, mat, "\n")

#Ejercicio 2
"Pidiendo el input del usuario pide dos números y crea una pequeña calculadora con los operadores básicos de suma, resta, multiplicación, división, y exponente."

n1 = int(input("Ingresa número 1\t"))
n2 = int(input("Ingresa número 2\t"))

print("Suma de los numeros:\t", n1 + n2)
print("Resta de los numeros:\t", n1 - n2)
print("Multiplicación de los numeros:\t", n1 * n2)
print("División de los numeros:\t", n1 / n2)
print("Exponente de los numeros:\t", n1 ** n2, "\n")

#Ejercicio 3
"Con loop while o for, realiza una lista de 10 numeros multiplos de 3, y después realiza una función de loop que sume todos los números dentro del arreglo."

a = 0

for i in range(1,11):
  i = i * 3
  a = a + i

print("Suma de los múltiplos de 3:\t", a, ("\n"))

#Ejercicio 4
"Con una función de if else, revisar si un número es par o es impar."
"Con una función de if else, revisar si un número es primo o no."

x = int(input("Ingrese número 1:\t"))
y = int(input("Ingrese número 2:\t"))

if x%2 == 0:
  print("El número ", x, " es par")
else:
  print("El número ", x, " es impar")

if y%2 == 0:
  print("El número ", y, " es par")
else:
  print("El número ", y, " es impar")

v1 = range(2,x)
v2 = range(2,y)
c = 0

for n in v1:
  if x % n == 0:
    c+=1

if (c > 0):
  print("El número", x, " no es primo.")
else:
  print("El número", x, "es primo.")

c = 0

for n in v2:
  if y % n == 0:
    c+=1

if (c > 0):
  print("El número", y, " no es primo.")
else:
  print("El número", y, "es primo.")

#Ejercicio 5
"\nUtilizando diferentes clases en python, crea una calculadora con los operadores básicos de suma, resta, multiplicación, división, y exponente."

class Calculadora:
  
  def __init__(self, x = 0.0, y = 0.0):
    self.x = float(x)
    self.y = float(y)
  
  def Suma(self):
    suma = self.x + self.y
    print("La suma es: \t", suma)

  def Resta(self):
    resta = self.x - self.y
    print("La resta es: \t", resta)

  def Multi(self):
    multi = self.x * self.y
    print("La mutiplicación es: \t", multi)

  def Div(self):
    div = self.x / self.y
    print("La división es: \t", div)

  def Exp(self):
    exp = self.x ** self.y
    print("El exponente es: \t", exp)

x = int(input("\nIngrese número 1:\t"))
y = int(input("Ingrese número 2:\t"))

calc = Calculadora(x, y)
calc.Suma()
calc.Resta()
calc.Multi()
calc.Div()
calc.Exp()

#Ejercicio 6
```
